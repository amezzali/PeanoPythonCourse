```python
# Programma che genera turni per le interrogazioni
import random
import math

elenco = sorted(['Barigozzi', 'Mandela', 'Tirubini', 'Burnic', 'Lillol', 'Trapezi', \
                'Durigol', 'Manetti', 'Caletto', 'Bomborni', 'Caffeo', 'Zoldi', 'Duribassi'])

elenco1 = elenco.copy()
num_turni = 3

print(math.ceil(len(elenco)/num_turni))

turni = []

for i in range(num_turni):
    
    try:
        aa = sorted(random.sample(elenco1,math.ceil(len(elenco)/num_turni)))
    except:
        aa = elenco1
        
    elenco1 = [x for x in elenco1 if x not in aa]
    turni.append(aa)
    

for i in range(len(turni)):
    print('\n- Turno ' + str(i+1))
    for i in turni[i]:
        print(i)
```

```python
# Programma per visualizzare l'andamento e la media dei voti

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
dictionary = {'Matematica': [7,4,6,7],\
             'Italiano': [4,8,3,2],\
             'Chimica': [7,7,7,8],\
             'Inglese': [4,5,2]}

colorlist = ['g','b','r','y']
j=0
for i in dictionary:
        
    color = colorlist[j]  
    plt.plot(range(len(dictionary[i])) ,dictionary[i],color)
    plt.plot(range(len(dictionary[i])),[np.mean(dictionary[i]) for x in range(len(dictionary[i]))]\
             , color+'--',label='_nolegend_')
    j += 1
    
plt.legend(dictionary.keys(),loc=(1.04,0.5))
plt.show()
```
