### Liste e grafici

Vediamo come creare dei grafici date delle liste di numeri. Utilizzeremo:

- liste
- la libreria *matplotlib*
- la libreria *numpy*

**Matplotlib** E' una libreria che permette di disegnare grafici. In questa sezione ci concentreremo su di un modulo particolare `matplotlib.pyplot`.

utilizzando le `magic function` di jupyter è possibile mostrare i plot direttamente nel jupyter notebook

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# creiamo una lista come abbiamo visto le scorse volte

x = []
for i in range(-10,10):
    x.append(i)

y = []
for i in range(-10,10):
    y.append(i**2)
    
print(x,'\n',y)

# un altro modo per creare liste in modo più veloce è il seguente

x = [i for i in range(-10,10)]
y = [i**2-2*i+1 for i in x]

# ora, disegnamo y in funzione di x

plt.plot(x,y)
plt.show()
```

Proviamo a inserire due funzioni nello stesso grafico.

Per utilizzare seni e coseni dobbiamo richiamare un'altra libreria. Scegliamo **numpy**, ma esistono alternative (ad esempio, la libreria **math**).

```python
import numpy as np

x = [i for i in range(-10,10)]

y_sin = np.sin(x)
y_cos = np.cos(x)

# due grafici
plt.plot(x, y_sin)
plt.plot(x, y_cos)

plt.xlabel('x axis label')
plt.ylabel('y axis label')

plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
```

Orrendo! la spaziatura (granularità) dell'ascissa è di una unità, quindi i punti sono troppo distanti e il grafico risultante è una spezzata.

il comando `range?` ci permette di vedere l'help

**Attenzione!** range(start, stop [,step]) 

- l'argomento *step* è opzionale
- possono solo essere interi! non posso usare un passo (*step*) decimale. 

Quindi, come posso creare una lista di numeri compresi tra -10 e 10 separati da 0.1 ?

**Varie soluzioni**

1. mi creo una funzione ad hoc
2. cerco ed utilizzo una funzione presente in un'altra libreria --> **np.arange**

```python
x = np.arange(0, 3 * np.pi, 0.1)

y_sin = np.sin(x)
y_cos = np.cos(x)

# due grafici
plt.plot(x, y_sin)
plt.plot(x, y_cos)

plt.xlabel('x axis label')
plt.ylabel('y axis label')

plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
```

Numpy permette un altro metodo per creare un: **np.linspace**

Notate la differenza tra linspace e arange?

```python
aa = np.linspace(0, 2*np.pi, 3)
bb = np.arange(0, 2*np.pi, 3)

print(aa)
print(bb)
```

Ora provo ad abbellire i miei grafici

```python
# con subplot creo un grfico composto un 
# pezzo alla volta

# x e y per il seno e il coseno
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# creo una griglia di subplot di altezza 2 e lunghezza 1,
# attivo il primo subplot
plt.subplot(2, 1, 1)

# disegno il primo grafico
plt.plot(x, y_sin)
plt.title('Sine')

# attivo il secondo grafico e lo disegno
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

# mostro la figura.
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
